```python
import openpyxl
import pandas as pd
import os
```

```python
excelname='raw_data'
sheetname='data'
wb=openpyxl.load_workbook(str(excelname)+'.xlsx')
ws=wb[str(sheetname)]
```

```python
dataset=[]
i=1
while(ws.cell(row=i,column=1).value!=None):
    row=[]
    j=1
    while(ws.cell(row=i,column=j).value!=None):
        row.append(ws.cell(row=i,column=j).value)
        j+=1
    dataset.append(row)
    i+=1
```

```python
trade = []
for i in range(0, len(dataset)):
    row = []
    for j in range(0, 11):
        row.append(0)
    trade.append(row)
trade[0][0] = "date"
trade[0][1] = "buy"
trade[0][2] = "hold"
trade[0][3] = "sell"
trade[0][4] = "number"
trade[0][5] = "value"
trade[0][6] = "period_return"
trade[0][7] = "closeprice"
trade[0][8] = "wma"
trade[0][9]  = "holding_days"
trade[0][10] = "Annualized Return"
pd.DataFrame(trade).head()
```

```python
for i in range(1, len(trade)):
    trade[i][0] = dataset[i][0]
    trade[i][7] = float(dataset[i][4])
```

```python
for i in range(3, len(trade)):
    temp =  trade[i][7]*3 + trade[i-1][7]*2 + trade[i-2][7]
    trade[i][8] = temp / 6
```

```python
pd.DataFrame(trade).head()
```

```python
for i in range(3, len(trade)):
    if(trade[i][7] > trade[i][8])and(trade[i-1][2]==1):
        trade[i][1] = 0
        trade[i][2] = 0
        trade[i][3] = 1
    elif(trade[i][7] < trade[i][8])and(trade[i-1][2]==0):
        trade[i][1] = 1
        trade[i][2] = 1
        trade[i][3] = 0
    elif(trade[i-1][2]>0):
        trade[i][1] = 0
        trade[i][2] = 1
        trade[i][3] = 0
```

```python
pd.DataFrame(trade).head(20)
```

```python
initial = 10000
for i in range(3, len(trade)):
    if(trade[i][1] == 1)and(trade[i][2]==1)and(trade[i][3] == 0):
        trade[i][4] = initial/trade[i][7]
        trade[i][5] = trade[i][4] * trade[i][7]
        trade[i][6] = round((trade[i][4] * trade[i][7] - initial)/initial, 4)
    elif(trade[i][1] == 0)and(trade[i][2]==0)and(trade[i][3] == 1):
        trade[i][4] = 0
        trade[i][5] = trade[i-1][4] * trade[i][7]
        trade[i][6] = round((trade[i-1][4] * trade[i][7] - initial)/initial, 4)
    elif(trade[i][1]==0)and(trade[i][2]==1)and(trade[i][3]==0):
        trade[i][4]=trade[i-1][4]
        trade[i][5]=trade[i][4]*trade[i][7]
        trade[i][6]=round((trade[i][7]*trade[i][4]-initial)/initial,4)
```

```python
pd.DataFrame(trade).head(20)
```

```python
#import datetime             
import datetime
from datetime import date
for i in range(1,len(trade)):
    trade[i][0]=datetime.datetime.strptime(trade[i][0],'%Y/%m/%d')
#holding_period
first=0
end=0
for i in range(1,len(trade)):
    if(trade[i][1]==1):
        trade[i][9]=1
        first=trade[i][0]
    elif((trade[i][2]>0)or(trade[i][3]==1)and(trade[i][1]!=1)):
        end=trade[i][0]
        trade[i][9]=(end-first).days+1
```

```python
pd.DataFrame(trade).head(30)
```

```python
#annual_return (holding_period)
for i in range(3,len(trade)):
    if(trade[i][3]!=0):
        trade[i][10]=(1+trade[i][6])**(365/trade[i][9])-1
```

```python
#annual_return (trading)
annual_return = 1
trading_counts = 0
for i in range(3, len(trade)):
    if(trade[i][3]==1):
        annual_return = annual_return * (1+trade[i][6])
        trading_counts = trading_counts + 1
```

```python
pd.DataFrame(trade).head(30)
```

```python
final_return = annual_return**(1/trading_counts)-1
```
